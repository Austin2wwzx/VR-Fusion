```python
import os
import os.path as osp
import pandas as pd
import numpy as np
import joblib

from natsort import natsorted
```

```python
LIDAR_ANNOTATION_PATH = '/Users/austin/Downloads/VRFusion/lidar/chengdu-2025-02-26/annotation'
file_path_list = natsorted(os.listdir(LIDAR_ANNOTATION_PATH))
```

```python
annotations_list = list()

for file_path in file_path_list:
    with open(osp.join(LIDAR_ANNOTATION_PATH, file_path), 'r') as f:
        annotations_list.extend(f.readlines())
```

```python
annotation_feature_list = list()

for annotation in annotations_list:
    annotation_list = annotation.replace('\n', '').split(',')
    
    annotation_feature_array = np.array([float(feature) for feature in annotation_list])
    annotation_feature_list.append(annotation_feature_array)

annotation_feature_array = np.array(annotation_feature_list)
feature_str_list = ['Timestamp', 'ID', 'Lon', 'Lat', 'x_size', 'y_size', 'z_size', 'Ori', 'x', 'y', 'z']
```

```python
df = pd.DataFrame(annotation_feature_array, columns=feature_str_list)
df['Timestamp'] = df['Timestamp'].astype(str)

for idx in range(len(annotations_list)):
    time_stamp = annotations_list[idx].replace('\n', '').split(',')[0][:-1]
    df.loc[idx, 'Timestamp'] = time_stamp

save_floder = '../../result/annotation'
# df.to_feather(osp.join(save_floder, '{}:lidar_annotation.feather').format(LIDAR_ANNOTATION_PATH.split('/')[-2]))
```
